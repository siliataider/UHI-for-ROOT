# A UHI version of hist019_TH2_projection.C

In this tutorial we will see how to convert the tutorial [hist019_TH2_projection.C](https://root.cern/doc/master/hist019__TH2__projection_8C.html) to Python using the UHI syntax.

```python
import matplotlib.pyplot as plt
import mplhep as hep
import numpy as np
import ROOT
import seaborn as sns
```

```python
h2 = ROOT.TH2F("h2", "", 40, -4, 4, 40, -20, 20)

h2[...] = np.histogram2d(
    np.random.normal(0, 1, 25000), 5*np.random.normal(0, 1, 25000), bins=(40, 40), range=[[-4, 4], [-20, 20]]
)[0]

projh2X = h2[::sum, :]
projh2Y = h2[:, ::sum]
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
sns.heatmap(h2.values(), ax=axes[1, 0])
hep.histplot(projh2X, ax=axes[1, 1], histtype="fill", color="purple", orientation="horizontal")
hep.histplot(projh2Y, ax=axes[0, 0], histtype="fill", color="purple")
axes[0, 1].text(0, 0.8, "This example demonstrates how to display\na histogram and its two projections.", fontsize=12)
axes[0, 1].axis("off")
plt.show()
```
